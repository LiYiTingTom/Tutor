```
def withoutPara(func):
  print(f"withoutPara({func}) --- start")

  def wrap():
    print(f"wrap() --- start")

    func()

    print(f"wrap() --- end")

  return wrap
```

```
def hello():
  print('hello world')

myFunc = withoutPara(hello)
myFunc()
```

## with parameters

```
def withPara(func):
  print(f"withPara({func}) --- start")

  def wrap(*argv):
    print(f"wrap({argv}) --- start")

    func(*argv)

    print(f"wrap({argv}) --- end")

  return wrap
```

```
def hi(name):
  print(f"hi {name}")
  

testFunc = withPara(hi)
testFunc("James")
```

# Syntax Candy


```
def codeBlock(func):
  warningC = '\033[93m'
  endC = '\033[0m'

  def wrap(*argv):
    print(f"{func.__name__} --- start {warningC}")
    func(*argv)
    print(f"{endC}{func.__name__} --- end")

  return wrap
```

```
@codeBlock
def warning(arg):
  print(arg)

@codeBlock
def foo(name):
  print(f"hello {name}")

testFunc = warning('This is a warning function test')
newFunc = foo('world')
```

## Multiple using

```
@withPara
@codeBlock
def multiUse(arg):
  print(arg)
```

## Syntax candy with parameters


```
def layer1(user_input):
  print(f"user_input = {user_input}")
  def decorator(func):
    warningC = '\033[93m'
    endC = '\033[0m'

    def wrap(*argv):
      print(f"{warningC}{func.__name__} --- start {endC}")
      func(*argv)
      print(f"{warningC}{func.__name__} --- end {endC}")
    
    return wrap
  return decorator
```

```
@layer1(user_input="hello World")
def myFunc(name):
  print(f"hello {name}")

myFunc('abc')
```
