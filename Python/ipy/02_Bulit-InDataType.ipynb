# Built-in Data Types

```
"""

# Numeric:
int, float, complex

# Text:
str

# Sequence:
list, tuple, range

# Mapping:
dict

# Set:
set, frozenset

# Boolean
bool

# Binary:
bytes, bytearray, memoryview

"""
```

```
# Numeric
INT = 20
FLOAT = 20.5
COMPLEX = -1j

# Text
STRING = "Hello World!"

#Sequence
LIST = [20, 20.5, -1j, "Hello World!"]
TUPLE = (1, 2, 3)
RANGE = range(0, 10)

# Mapping
DICTIONARY = {'name':"James", 'class':"Python"}

# Set
SET = {"apple", "banana", "apple"}
FORZENSET = ({"apple", "banana", "apple"})

# Boolean
BOOL = False

# Byte
BYTE = b"Hello"
BYTEARRAY = bytearray(5)
MEMORYVIEW = memoryview(bytes(5))
```

```
# Force transform data type
a = 1.32
b = str(a)
print(b + str(1.3))
```
