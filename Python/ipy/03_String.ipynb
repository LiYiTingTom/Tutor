- Slicing

```
str1 = "0123456789"
print(str1[2:5])

print(str1[-5:-2])

print(str1[:])

print(str1[:-5])

print(str1[-3:])
```

- Length

```
print(len(str2))
```

- Methods

```
# strip() : remove whitespace for the bigining or the end
a = "  Hello, World !  "
print("~%s~" %(a.strip()))

# lower() : return lower case
print("~%s~" %(a.lower()))
# upper() : return upper case
print("~%s~" %(a.upper()))

# replace() : replace with sub-string
print("~%s~" %(a.replace("World", "Ketty")))

# split() : split string into multiple sub-string
print("~%s~"%a.split(','))
```

- Check

```
txt = "The rain in Spain stays mainly in the plain"

print("ain" in txt)
print("ain" not in txt)
```

- Concatenation

```
a = "Hello"
print(a)
a += " World"
print(a)
```
