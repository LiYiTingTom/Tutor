````
class Human():
  def __init__(self, age, name, sex):
    # public
    self.sex = sex
    # protected
    self._age = age
    # private
    self.__name = name
  
  def __gotoBathroom(self):
    print(f"{self.__name} goto Bathroom")
  
  def eat(self, food):
    print(f"{self.__name} eat {food}")
  
  def getName(self):
    return self.__name
  
  def whereAreYou(self):
    self.__gotoBathroom()

````

````
class Animal():
  """ a testing class """
  totalNum = 0
  """
    Constructor
  """
  def __init__(self, sex, name, age):
    # static attribute
    Animal.totalNum += 1

    # public attribute
    self.sex = sex

    # protected attribute
    self.name = name

    # private attribute
    self.__age = age

  def have_some_food(self):
    print(f"{self.name} had food.")

  """
    set, get, del methods

    @property
    def name(self):
      return self._name

    @name.setter
    def name(self, newName):
      self._name = newName
    
    @name.deleter
    def name(self):
      del self._name

  """
  @property
  def age(self):
    print(f"get : age = {self.__age}")
    return self.__age
""" a testing class """""" a testing class """
  @age.setter
  def age(self, age):
    print(f"set : age = {age}")
    self.__age = age

  @age.deleter
  def age(self):
    print(f"del : age")
    del self.__age
````

# Inheritance

````
class Human(Animal):  # inheritance Animal class
  totalNum = 0

  """
    Constractor
  """
  def __init__(self, sex, name, age, country):
    Human.totalNum += 1

    # call parent's contructor
    super().__init__(sex, name, age)
    
    # self attribute
    self._country = country
  
  """
    @property
  """
  @property
  def country(self):
    return self._country
  
  @country.setter
  def country(self, country):
    self._country = country
  
  @country.deleter
  def country(self):
    del self._country

````

``` bash
# PL
<attr> var_
<attr> flow_controller_
  ## OOP
  <attr> object_
  <method> print("Polymorphism")
  <method> print("Inheritance")
    ### Python
    <attr> dict_
    <method> print("Lambda")
```

````
class PL():
  def __init__(var_, flow_controller_):
    self.var_ = var_
    self.flow_controller_ = flow_controller_

class OOP(PL):
  def __init__(var_, flow_controller_, object_):
    self.object_ = object_
    super().__init__(var_, flow_controller_)
  
  def polymorphism():
    print("polymorphism")
  def inheritance():
    print("inheritance")

class Python(OOP):
  def __init__(var_, flow_controller_, object_, dict_):
    self.dict_ = dict_
    super().__init__(object_)
    super().super().__init__(var_, flow_controller_)
  
  def Lambda():
    print("Lambda")
````
