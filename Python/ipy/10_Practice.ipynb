```
import matplotlib.pyplot as plt


def r(n):
  return 1 if n == 1 else 2*r(n-1)

def g(n):
  return 20 if n == 1 else 2*(g(n-1) - r(n-1))

reds = []
greens = []

for n in range(1, 100):
  reds.append(r(n))
  greens.append(g(n))
  
  if g(n) == 0:
    print(f"{n} th generation")
    break

plt.plot(range(1, n+1), reds , 'red')
plt.plot(range(1, n+1), greens, 'green')
plt.legend()
```

```
def triangle(h):
  for i in range(1, h+1):
    for j in range(1, i+1):
      print("%4d"%(i*j), end='')

    print()

triangle(30)
```

```
def compute(a, b):
  answer = 0
  for i in range(a, b+1):
    answer += i
  return answer


def super_compute(a, b):
  return sum(range(a, b+1))
```

```
def max_numbers(*args):

  return (sorted(*args, reverse=True)[:3])


print(max_numbers([1,5,1234,15,25,3456,36,546,54,7,34,345,123]))
```
