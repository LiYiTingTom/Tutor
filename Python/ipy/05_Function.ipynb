````
def gcd(m, n):
  if n == 0: return m
  else: return gcd(n, m%n)

print(gcd(1234134, 12342))
````

````
# Find minimum by function
"""
  using API (application programming interface):

    def myMin(*argv):
      ...
    return minimum
"""

def myMin(*argv):
  mini = argv[0]
  for arg in argv:
    if mini > arg:
      mini = arg
  return mini

number = (1, 2, 3, 1, 6, 5, 2, 0, 1, 3, 6, 2, 3, 7, 8)
print(f"The minimum in number list is: {myMin(*number)}")
````

````
array = (3,2,3,4,5,2,2,314,5,12345,234,5,436,54,6)


def myMin(array):
  mini = array[0]
  for i in array:
    if i < mini:
      mini = i
  return mini

print(f"myMin(array) = {myMin(array)}")
min(array)
````

````
import math

def prepareFact(max):
  prime = [1] * max
  for i in range(2, int(math.sqrt(max))):
    if prime[i] == 1:
      for j in range(2*i, max):
        if j % i == 0: prime[j] = 0
  primes = [i for i in range(2, max) if prime[i] == 1]

  def factor(num):
    nums = []
    i = 0
    while primes[i] ** 2 <= num:
      if num % primes[i] == 0:
        nums.append(primes[i])
        num //= primes[i]
      else:
        i += 1
    nums.append(num)
    f = [0] * len(nums)
    for i in range(len(f)):
      f[i] = nums[i]
    return f

  return factor

factor = prepareFact(1000)
print(factor(100))
print(factor(500))
print(factor(900))
````
