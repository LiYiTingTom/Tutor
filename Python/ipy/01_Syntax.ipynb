```
x = "testing"

def funct():
  # Same output
#  print("This's a %s"%x)
  print("This's a "+x)

  # Using 'global' symbol to create a new global var
  global y
  y = "another testing"

def funct2():
  # Using global var y
  print("This's a "+y)

funct2()
funct()
```
