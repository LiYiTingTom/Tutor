# 作業一 - 九九乘法表

1. 使用函式
> ``` python 
> def table():
>    print(...)
>    ...
> ```

````
def table():
  """ which can print 1x1 to 9x9 """
  for i in range(1, 10):
    for j in range(1, 10):
      print(f"{i} x {j} = {i*j}")


# Answer
print(table())

"""
1 x 1 = 1
1 x 2 = 2
....
9 x 8 = 72
9 x 9 = 81
"""
````

# 作業二 - BMI計算機

- 公式為:體重(公斤)/(身高(公尺))**2

1. 使用函式
> ``` python
> def bmi(weight, height):
>     ...
>     return bmi_value
> ```

````
def bmi(weight, height):

  if height > 3:
    height = height * 0.01

  bmi_value = weight/(height**2)

  if bmi_value < 18.5:
    print("過輕")
  elif 18.5 <= bmi_value < 24:
    print("健康")
  elif 24 <= bmi_value < 27:
    print("過重")
  elif bmi_value >= 27:
    print("肥胖")

  return bmi_value

# Answer
print(bmi(, 170))

"""
19.0
"""
````
